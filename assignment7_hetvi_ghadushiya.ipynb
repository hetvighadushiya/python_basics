a. Differences in np.array, np.asarray, and np.asanyarray
np.array():

This function creates a new NumPy array. It copies the input data into a new array, regardless of the input type (list, tuple, or another array).
Example:

```python
import numpy as np
a = np.array([1, 2, 3])
```

Characteristics: Creates a new array and makes a copy of the data.

np.asarray():

This function also creates an array, but it does not make a copy if the input is already an ndarray. It is used to convert input data to an array, but if the input is already an array, it returns the original array.
Example:

Characteristics: Returns the original array if the input is already an array, saving memory

np.asanyarray():

Similar to np.asarray(), but it allows input data to be of any type that can be converted to an array (including subclasses like np.matrix).

```python
c = np.asanyarray([[1, 2], [3, 4]], dtype=np.matrix)
```

Characteristics: Returns a base array if possible, but will return subclasses if the input is a subclass.

b. Difference Between Deep Copy and Shallow Copy
Shallow Copy:

A shallow copy creates a new object, but it does not recursively copy objects contained within the original object. Instead, it copies references to those objects.
Changes to mutable objects in the original will be reflected in the shallow copy.
Example :

```python
import copy
original = [1, 2, [3, 4]]
shallow = copy.copy(original)
shallow[2][0] = 'Changed'
print(original)
```

Deep Copy:

A deep copy creates a new object and recursively copies all objects contained within the original object, meaning changes to the original do not affect the deep copy.
Example:

```python
deep = copy.deepcopy(original)
deep[2][0] = 'Deep Change'
print(original) 
print(deep)     
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

date_range = pd.date_range(start='2024-01-01', periods=100, freq='D')

np.random.seed(0) 
temperature = np.random.uniform(low=15, high=30, size=len(date_range))  # Temperature in Celsius
humidity = np.random.uniform(low=30, high=90, size=len(date_range))  # Humidity in %

data = {
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
}
df = pd.DataFrame(data)

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature', marker='o', linestyle='-')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity', marker='x', linestyle='--')
ax2.tick_params(axis='y', labelcolor='tab:blue')

plt.title('Temperature and Humidity Over Time')
fig.tight_layout() 

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0) 
data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(12, 6))
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.5, color='g', edgecolor='black', label='Histogram')

x = np.linspace(min(data), max(data), 100)
pdf = (1 / (np.sqrt(2 * np.pi))) * np.exp(-0.5 * (x ** 2))

plt.plot(x, pdf, color='red', linewidth=2, label='PDF')

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)  
x = np.random.rand(100) * 10 
y = np.random.rand(100) * 10 

colors = []
for i in range(len(x)):
    if x[i] >= 5 and y[i] >= 5:
        colors.append('red')      
    elif x[i] < 5 and y[i] >= 5:
        colors.append('blue')    
    elif x[i] < 5 and y[i] < 5:
        colors.append('green')    
    else:
        colors.append('orange')  

# Step 3: Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=colors)

plt.title('Quadrant-wise Scatter Plot') 
plt.xlabel('X-axis')                      
plt.ylabel('Y-axis')                    

red_patch = plt.Line2D([0], [0], marker='o', color='w', label='Quadrant 1', markerfacecolor='red', markersize=10)
blue_patch = plt.Line2D([0], [0], marker='o', color='w', label='Quadrant 2', markerfacecolor='blue', markersize=10)
green_patch = plt.Line2D([0], [0], marker='o', color='w', label='Quadrant 3', markerfacecolor='green', markersize=10)
orange_patch = plt.Line2D([0], [0], marker='o', color='w', label='Quadrant 4', markerfacecolor='orange', markersize=10)

plt.legend(handles=[red_patch, blue_patch, green_patch, orange_patch], title='Quadrants')

plt.show()
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

x = np.linspace(0, 2 * np.pi, 100) 
y = np.sin(x)

output_notebook() 
p = figure(title='Sine Wave Function', x_axis_label='X', y_axis_label='sin(X)', width=800, height=400)

p.line(x, y, line_width=2, color='blue')

p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5

show(p)
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

categories = [f'Category {i}' for i in range(1, 11)]
values = np.random.randint(1, 100, size=len(categories))

data = pd.DataFrame({'Category': categories, 'Value': values})

source = ColumnDataSource(data)

output_notebook()  
p = figure(x_range=data['Category'], title='Random Categorical Bar Chart', 
           x_axis_label='Categories', y_axis_label='Values', 
           height=400, width=800)

p.vbar(x='Category', top='Value', source=source, 
        fill_color='blue', line_color='white', 
        width=0.8)

hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

show(p)
```
