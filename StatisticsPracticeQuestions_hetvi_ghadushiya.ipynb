## Q1. Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.
After generating the list, find the following:

  (i) Write a Python function to calculate the mean of a given list of numbers.Create a function to find the median of a list of numbers.
  (ii) Develop a program to compute the mode of a list of integers.
  (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
  (iv) Write a Python function to find the geometric mean of a list of positive numbers.
  (v) Create a program to calculate the harmonic mean of a list of values.
  (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
  (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of
outliers.

````python
# Q1: Generate a list of 100 integers between 90 to 130
import random
import math
from collections import Counter

# Generate the list
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Functions for Mean and Median
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    sorted_list = sorted(numbers)
    n = len(sorted_list)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_list[mid - 1] + sorted_list[mid]) / 2
    return sorted_list[mid]

# (ii) Compute the Mode
def calculate_mode(numbers):
    freq = Counter(numbers)
    max_freq = max(freq.values())
    return [k for k, v in freq.items() if v == max_freq]

# (iii) Weighted Mean
def calculate_weighted_mean(values, weights):
    return sum(v * w for v, w in zip(values, weights)) / sum(weights)

# (iv) Geometric Mean
def calculate_geometric_mean(numbers):
    product = math.prod(numbers)
    return product ** (1 / len(numbers))

# (v) Harmonic Mean
def calculate_harmonic_mean(numbers):
    return len(numbers) / sum(1 / x for x in numbers)

# (vi) Midrange
def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# (vii) Trimmed Mean
def calculate_trimmed_mean(numbers, trim_percent):
    sorted_list = sorted(numbers)
    trim_count = int(len(sorted_list) * trim_percent / 100)
    trimmed_list = sorted_list[trim_count:-trim_count]
    return calculate_mean(trimmed_list)

print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))
print("Weighted Mean:", calculate_weighted_mean(int_list, [1]*100))
print("Geometric Mean:", calculate_geometric_mean([x for x in int_list if x > 0]))
print("Harmonic Mean:", calculate_harmonic_mean([x for x in int_list if x > 0]))
print("Midrange:", calculate_midrange(int_list))
print("Trimmed Mean (10%):", calculate_trimmed_mean(int_list, 10))
````

## Q2. Generate a list of 500 integers containing values between 200 to 300 and store it in the variable `int_list2`.
After generating the list, find the following:
  (i) Compare the given list of visualization for the given data:
    1. Frequency & Gaussian distribution
    2. Frequency smoothened KDE plot
    3. Gaussian distribution & smoothened KDE plot
  (ii) Write a Python function to calculate the range of a given list of numbers.
  (iii) Create a program to find the variance and standard deviation of a list of numbers.
  (iv) Implement a function to compute the interquartile range (IQR) of a list of values.
  (v) Build a program to calculate the coefficient of variation for a dataset.
  (vi) Write a Python function to find the mean absolute deviation (MAD) of a list of numbers.
  (vii) Create a program to calculate the quartile deviation of a list of values.
  (viii) Implement a function to find the range-based coefficient of dispersion for a dataset

````python
# Q2: Generate a list of 500 integers between 200 to 300
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate the list
int_list2 = [random.randint(200, 300) for _ in range(500)]

# (i) Visualizations
def visualize_data(data):
    plt.figure(figsize=(15, 5))

    # 1. Frequency & Gaussian distribution
    plt.subplot(1, 3, 1)
    sns.histplot(data, kde=False, color='blue', bins=20, label="Frequency")
    x = np.linspace(min(data), max(data), 1000)
    plt.plot(x, norm.pdf(x, np.mean(data), np.std(data)) * len(data), 'r', label="Gaussian")
    plt.title("Frequency & Gaussian Distribution")
    plt.legend()

    # 2. Frequency smoothened KDE plot
    plt.subplot(1, 3, 2)
    sns.histplot(data, kde=True, color='green', bins=20)
    plt.title("Frequency Smoothened KDE Plot")

    # 3. Gaussian distribution & KDE plot
    plt.subplot(1, 3, 3)
    sns.kdeplot(data, color='purple', label="KDE")
    plt.plot(x, norm.pdf(x, np.mean(data), np.std(data)), 'r--', label="Gaussian")
    plt.title("Gaussian & KDE Plot")
    plt.legend()

    plt.tight_layout()
    plt.show()

visualize_data(int_list2)

# (ii) Range
def calculate_range(numbers):
    return max(numbers) - min(numbers)

# (iii) Variance and Standard Deviation
def calculate_variance(numbers):
    mean = sum(numbers) / len(numbers)
    return sum((x - mean) ** 2 for x in numbers) / len(numbers)

def calculate_std_dev(numbers):
    return calculate_variance(numbers) ** 0.5

# (iv) Interquartile Range (IQR)
def calculate_iqr(numbers):
    q1 = np.percentile(numbers, 25)
    q3 = np.percentile(numbers, 75)
    return q3 - q1

# (v) Coefficient of Variation
def calculate_coefficient_of_variation(numbers):
    return (calculate_std_dev(numbers) / calculate_mean(numbers)) * 100

# (vi) Mean Absolute Deviation (MAD)
def calculate_mad(numbers):
    mean = calculate_mean(numbers)
    return sum(abs(x - mean) for x in numbers) / len(numbers)

# (vii) Quartile Deviation
def calculate_quartile_deviation(numbers):
    return calculate_iqr(numbers) / 2

# (viii) Range-based Coefficient of Dispersion
def calculate_range_dispersion(numbers):
    range_val = calculate_range(numbers)
    mean = calculate_mean(numbers)
    return range_val / mean

print("Range:", calculate_range(int_list2))
print("Variance:", calculate_variance(int_list2))
print("Standard Deviation:", calculate_std_dev(int_list2))
print("Interquartile Range (IQR):", calculate_iqr(int_list2))
print("Coefficient of Variation (%):", calculate_coefficient_of_variation(int_list2))
print("Mean Absolute Deviation (MAD):", calculate_mad(int_list2))
print("Quartile Deviation:", calculate_quartile_deviation(int_list2))
print("Range-based Coefficient of Dispersion:", calculate_range_dispersion(int_list2))
````

## Q3. Write a Python class representing a discrete random variable with methods to calculate its expected value and variance

````python
class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
        """
        Initialize the discrete random variable.
        :param outcomes: A list of outcomes (x_i).
        :param probabilities: A list of probabilities (p_i) corresponding to the outcomes.
        """
        if len(outcomes) != len(probabilities):
            raise ValueError("Outcomes and probabilities must have the same length.")
        if not math.isclose(sum(probabilities), 1.0):
            raise ValueError("Probabilities must sum to 1.")
        
        self.outcomes = outcomes
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculate the expected value E(X).
        :return: The expected value.
        """
        return sum(x * p for x, p in zip(self.outcomes, self.probabilities))

    def variance(self):
        """
        Calculate the variance Var(X).
        :return: The variance.
        """
        mean = self.expected_value()
        return sum(p * (x - mean) ** 2 for x, p in zip(self.outcomes, self.probabilities))

# Example usage
outcomes = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]

# Create an instance of the random variable
rv = DiscreteRandomVariable(outcomes, probabilities)

# Calculate expected value and variance
print("Expected Value:", rv.expected_value()) 
print("Variance:", rv.variance())             
````

## Q4.Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.

## Q5.Create a Python function to generate random samples from a given probability distribution (e.g.,binomial, Poisson) and calculate their mean and variance.

````python
import numpy as np

def generate_samples_and_stats(distribution, params, sample_size=1000):
    """
    Generate random samples from a specified probability distribution
    and calculate their mean and variance.

    :param distribution: The distribution name ('binomial', 'poisson', etc.)
    :param params: Parameters for the distribution (e.g., n, p for binomial)
    :param sample_size: Number of samples to generate (default: 1000)
    :return: A tuple containing (mean, variance)
    """
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, sample_size)
    elif distribution == 'poisson':
        lam = params[0]
        samples = np.random.poisson(lam, sample_size)
    else:
        raise ValueError(f"Unsupported distribution: {distribution}")

    mean = np.mean(samples)
    variance = np.var(samples)

    return mean, variance

binomial_mean, binomial_variance = generate_samples_and_stats('binomial', (10, 0.5))
print("Binomial Distribution (n=10, p=0.5):")
print(f"Mean: {binomial_mean}, Variance: {binomial_variance}")

poisson_mean, poisson_variance = generate_samples_and_stats('poisson', (4,))
print("\nPoisson Distribution (lambda=4):")
print(f"Mean: {poisson_mean}, Variance: {poisson_variance}")
````

## Q6.Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples.

````python
import numpy as np

def gaussian_statistics(mean, std_dev, sample_size):
    """
    Generate random samples from a Gaussian distribution and compute statistics.

    :param mean: Mean of the Gaussian distribution (mu)
    :param std_dev: Standard deviation of the Gaussian distribution (sigma)
    :param sample_size: Number of random samples to generate
    :return: A tuple containing (mean, variance, standard deviation) of the samples
    """
    samples = np.random.normal(mean, std_dev, sample_size)
    
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)
    
    return sample_mean, sample_variance, sample_std_dev

mean = 10   
std_dev = 2     
sample_size = 1000 

sample_mean, sample_variance, sample_std_dev = gaussian_statistics(mean, std_dev, sample_size)

print(f"Generated {sample_size} samples from N({mean}, {std_dev}^2):")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Variance: {sample_variance:.2f}")
print(f"Sample Standard Deviation: {sample_std_dev:.2f}")
````

````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Load the tips dataset
tips = sns.load_dataset('tips')

# Extract the columns for analysis
total_bill = tips['total_bill']
tip = tips['tip']

# (i) Function to calculate skewness
def calculate_skewness(data):
    return skew(data)

# (ii) Determine the type of skewness
def skewness_type(data):
    skewness_value = calculate_skewness(data)
    if skewness_value > 0:
        return "Positive Skewness"
    elif skewness_value < 0:
        return "Negative Skewness"
    else:
        return "Approximately Symmetric"

# (iii) Function to calculate covariance
def calculate_covariance(col1, col2):
    return np.cov(col1, col2, ddof=0)[0, 1]

# (iv) Function to calculate Pearson correlation coefficient
def calculate_correlation(col1, col2):
    return np.corrcoef(col1, col2)[0, 1]

# (v) Visualize correlation using scatter plot
def visualize_correlation(col1, col2, col1_name, col2_name):
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=col1, y=col2)
    plt.title(f"Scatter Plot: {col1_name} vs {col2_name}")
    plt.xlabel(col1_name)
    plt.ylabel(col2_name)
    plt.grid(True)
    plt.show()

# Results
print("(i) Skewness:")
print(f"  Skewness of total_bill: {calculate_skewness(total_bill):.2f}")
print(f"  Skewness of tip: {calculate_skewness(tip):.2f}")

print("\n(ii) Type of Skewness:")
print(f"  Total_bill: {skewness_type(total_bill)}")
print(f"  Tip: {skewness_type(tip)}")

print("\n(iii) Covariance:")
print(f"  Covariance between total_bill and tip: {calculate_covariance(total_bill, tip):.2f}")

print("\n(iv) Pearson Correlation Coefficient:")
print(f"  Correlation between total_bill and tip: {calculate_correlation(total_bill, tip):.2f}")

print("\n(v) Visualization:")
visualize_correlation(total_bill, tip, "Total Bill", "Tip")
````

## Q10.Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

````python
import math

def poisson_pmf(k, lambda_rate):
    """
    Calculate the PMF of a Poisson distribution.

    Parameters:
    k (int): The number of occurrences (non-negative integer).
    lambda_rate (float): The average rate of occurrences ().

    Returns:
    float: The probability of observing k occurrences.
    """
    if k < 0:
        return 0 
    return (math.exp(-lambda_rate) * (lambda_rate ** k)) / math.factorial(k)

lambda_rate = 4.0
k = 3             

pmf_value = poisson_pmf(k, lambda_rate)
print(f"The PMF of Poisson distribution for k={k} and ={lambda_rate} is {pmf_value:.5f}")
````

## Q11.A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors
who make a purchase). They collect data from the old and new layouts to compare.
To generate the data use the following command:
```python
import numpy as np
# 50 purchases out of 1000 visitors
old_layout = np.array([1] * 50 + [0] * 950)
# 70 purchases out of 1000 visitors  
new_layout = np.array([1] * 70 + [0] * 930)
  ```
Apply z-test to find which layout is successful.

````python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Generate data
old_layout = np.array([1] * 50 + [0] * 950) 
new_layout = np.array([1] * 70 + [0] * 930)  

successes = [old_layout.sum(), new_layout.sum()]  
n_obs = [len(old_layout), len(new_layout)]       

z_stat, p_value = proportions_ztest(successes, n_obs, alternative='larger')

# Print results
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The new layout leads to a significantly higher conversion rate.")
else:
    print("Fail to reject the null hypothesis: No significant difference in conversion rates.")
````

## Q15.U A company is testing two different website layouts to see which one leads to higher click-through rates.
Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
freedom, and p-value.
Use the following data:
```python

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]

layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

````python
import numpy as np
from scipy import stats

# Data for layout A and B
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

mean_a = np.mean(layout_a_clicks)
mean_b = np.mean(layout_b_clicks)

var_a = np.var(layout_a_clicks, ddof=1) 
var_b = np.var(layout_b_clicks, ddof=1)  

n_a = len(layout_a_clicks)
n_b = len(layout_b_clicks)

t_stat = (mean_a - mean_b) / np.sqrt((var_a / n_a) + (var_b / n_b))

df = ((var_a / n_a + var_b / n_b) ** 2) / \
    (( (var_a / n_a) ** 2 / (n_a - 1) ) + ((var_b / n_b) ** 2 / (n_b - 1)))

p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

print(f"Mean of layout A: {mean_a:.2f}")
print(f"Mean of layout B: {mean_b:.2f}")
print(f"t-statistic: {t_stat:.2f}")
print(f"Degrees of freedom: {df:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two layouts.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the two layouts.")
````

```python

existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]

````python
import numpy as np
from scipy import stats

existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

mean_existing = np.mean(existing_drug_levels)
mean_new = np.mean(new_drug_levels)

var_existing = np.var(existing_drug_levels, ddof=1)  
var_new = np.var(new_drug_levels, ddof=1) 

n_existing = len(existing_drug_levels)
n_new = len(new_drug_levels)

t_stat = (mean_existing - mean_new) / np.sqrt((var_existing / n_existing) + (var_new / n_new))

df = ((var_existing / n_existing + var_new / n_new) ** 2) / \
    (( (var_existing / n_existing) ** 2 / (n_existing - 1) ) + ((var_new / n_new) ** 2 / (n_new - 1)))

p_value = stats.t.cdf(t_stat, df)

print(f"Mean of existing drug group: {mean_existing:.2f}")
print(f"Mean of new drug group: {mean_new:.2f}")
print(f"t-statistic: {t_stat:.2f}")
print(f"Degrees of freedom: {df:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The new drug is more effective in reducing cholesterol.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the new drug is more effective.")
````

````python
import numpy as np
from scipy import stats

np.random.seed(0)  
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

mean_male = np.mean(male_salaries)
mean_female = np.mean(female_salaries)
var_male = np.var(male_salaries, ddof=1) 
var_female = np.var(female_salaries, ddof=1) 
n_male = len(male_salaries)
n_female = len(female_salaries)

t_stat = (mean_male - mean_female) / np.sqrt((var_male / n_male) + (var_female / n_female))

df = ((var_male / n_male + var_female / n_female) ** 2) / \
     (((var_male / n_male) ** 2) / (n_male - 1) + ((var_female / n_female) ** 2) / (n_female - 1))

p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

print(f"Mean male salary: ${mean_male:.2f}")
print(f"Mean female salary: ${mean_female:.2f}")
print(f"t-statistic: {t_stat:.2f}")
print(f"Degrees of freedom: {df:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a statistically significant difference in salaries.")
else:
    print("Fail to reject the null hypothesis: No statistically significant difference in salaries.")
````

## Q21.A political analyst wants to determine if there is a significant association between age groups and voter preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify them into different age groups and candidate preferences. Perform a Chi-Square test to determine if there is a significant association between age groups and voter preferences.
Use the below code to generate data:
```python

age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

````python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_stat)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)

alpha = 0.05
if p_value < alpha:
    print("\nThere is a significant association between age groups and voter preferences.")
else:
    print("\nThere is no significant association between age groups and voter preferences.")
````
